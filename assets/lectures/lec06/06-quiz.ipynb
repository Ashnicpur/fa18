```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

plt.rcParams['figure.figsize'] = (14, 7)
```

```python
data12 = pd.read_csv('data/DatasaurusDozen.tsv', sep='\t')
data12.head()
```

```python
data = data12.query("dataset in ['dino', 'slant_up']")
data.head()
```

```python
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=data,
           col_wrap=2, ci=None, height=6, fit_reg=False);
```

```python
import numpy as np

x = data.query("dataset == 'dino'")['x']
y1 = ((1-(x-x.mean())**2))/20 + 100 + np.random.normal(0, 3, x.shape)
y2 = x + np.random.normal(0, 3, x.shape)

dataa = pd.DataFrame({'dataset': 'I', 'x': x, 'y': y1})
datab = pd.DataFrame({'dataset': 'II', 'x': x, 'y': y2})
data2 = pd.concat([dataa, datab])

sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=data2,
           col_wrap=4, ci=None, height=6, fit_reg=False);
```
